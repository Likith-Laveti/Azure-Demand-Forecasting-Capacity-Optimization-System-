```python
supply_path = "abfss://destination@storagedatafactory2605.dfs.core.windows.net/DataCoSupplyChainDataset/DataCoSupplyChainDataset_Cleaned_Final_Unique.csv"

df_supply_raw = (
    spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(supply_path)
)
```

```python
import re

df_supply_clean_cols = df_supply_raw

for col in df_supply_raw.columns:
    clean_col = col.lower()
    clean_col = clean_col.replace(" ", "_")
    clean_col = re.sub(r"[(){};,=\n\t]", "", clean_col)
    df_supply_clean_cols = df_supply_clean_cols.withColumnRenamed(col, clean_col)

df_supply_clean_cols.printSchema()
```

```python
bronze_supply_path = "abfss://fbronze@storagedatafactory2605.dfs.core.windows.net/supply_chain_delta/"

(df_supply_clean_cols.write
    .format("delta")
    .mode("overwrite")
    .save(bronze_supply_path)
)
```
